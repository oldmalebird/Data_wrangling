```python
from bs4 import BeautifulSoup
import os
import pandas as pd
```

```python
# 字典列表可以逐个创建文件，随后转化为一个 DataFrame
df_list = []
folder = 'rt_html'
for movie_html in os.listdir(folder):
    with open(os.path.join(folder, movie_html),encoding='utf-8') as file:
        # open时如果不加encoding='utf-8'，会出现'gbk' codec can't decode byte 0x93 in position 
        # 正确操作需要花费 ~15 秒运行
        soup = BeautifulSoup(file, 'lxml')
        title = soup.find('title').contents[0][:-len(' - Rotten Tomatoes')]
        #print(title)
        audience_score = soup.find('div',class_='audience-score meter').find('span').contents[0][:-1]
        #class后面有下划线，因为class是保留字
        #print(audience_score)
        num_of_audience_ratings = soup.find('div',class_='audience-info hidden-xs superPageFontColor')
        num_of_audience_ratings = num_of_audience_ratings.find_all('div')[1].contents[2].strip().replace(',','')
        #print(num_of_audience_ratings)
        df_list.append({'title': title,
'audience_score': int(audience_score),
'number_of_audience_ratings': int(num_of_audience_ratings)})
df = pd.DataFrame(df_list, columns = ['title', 'audience_score', 'number_of_audience_ratings'])
```
